- 缺失值

```python
dict3 = {}
dict4 = {}
np.random.seed(0)
rdi = np.random.randint(10,100,(10))
for di in rdi:
    if di%2 == 0:
        dict3[di] = di//10
    else:
        dict4[di] = di%10
df2 = pd.DataFrame({'十位':pd.Series(dict3),'digit':pd.Series(dict4)})
df2
```

```python
#对列加总
df2['digit'].sum()
```

```python
#缺失值判断
df2.isnull()              #df2.notnull()非缺失值判断
```

```python
#删除含有缺失值的样本
df2.dropna()          #df2不变
```

```python
#填补
df2.fillna(-1)
#df2.fillna(method = 'ffill')默认用前面（往列上看）的值插补
#df2.fillna(method = 'bfill')默认用后面的值插补
#df2.fillna(method = 'bfill',axis=1)
```

## 3. 表连接

- concat 增加行

```python
dict1 = {}
dict2 = {}
np.random.seed(0)
rdi = np.random.randint(10,100,(10))
print(rdi)
for li in rdi:
    dict1[li] = li//10
    dict2[li] = li%10
sr_10 = pd.Series(dict1)
sr_1 = pd.Series(dict2)
df1 = pd.DataFrame({'十位':sr_10,'digit':sr_1})
df1
dict3 = {}
dict4 = {}
dict5 = {}
np.random.seed(0)
rdi2 = np.random.randint(100,900,(10))
for di in rdi2:
    dict3[di] = di//100
    dict4[di] = di%100//10
    dict5[di] = di%10
df3 = pd.DataFrame({'百位':pd.Series(dict3),'十位':pd.Series(dict4),'digit':pd.Series(dict5)})
print(df1);print(df3)
```

```python
pd.concat([df1,df3],join='inner')
```

```python
pd.concat([df1,df3],join='outer')
```

- merge增加列

```python
df4 = pd.DataFrame({'十位':pd.Series(dict4),'digit':pd.Series(dict5)})
df5 = pd.DataFrame({'十位':pd.Series(dict4),'百位':pd.Series(dict3)});print(df4);print(df5)
```

```python
pd.merge(df4,df5)#只要十位有的 全部展示
```

```python
pd.merge(df1,df3,on='digit',how='outer')
```

- groupby

```python
df6 = pd.merge(df4,df5)
df6
```

```python
df6.groupby('digit').sum()  #bysort
```
